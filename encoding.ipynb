#Encding

##Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data = ['merah','kuning','hijau','kuning']
label_encoder = LabelEncoder()
num_data = label_encoder.fit_transform(data)
num_data
```

##One-Hot-Encoding

```python
import pandas as pd
```

```python
data = ['merah','kuning','hijau','kuning']
one_hot_encoder = pd.get_dummies(data,dtype=int)
one_hot_encoder
```

##Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = [['low'],['jawa'],['jawa'],['medium']]
ordinal_encoder = OrdinalEncoder()
num_data = ordinal_encoder.fit_transform(data)
num_data
```

##Target Encoding

```python
data = pd.DataFrame({
    'city' : ['jawa','batak','jawa','jawa','batak'],
    'target':[1,0,1,0,1]
    })
mean_target = data.groupby('city')['target'].mean()
data['city'] = data['city'].map(mean_target)
data
```
